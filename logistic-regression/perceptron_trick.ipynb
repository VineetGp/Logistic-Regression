```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_classification(n_samples=100,n_features=2,n_classes=2,class_sep=15,n_redundant=0,
                          hypercube=False,n_informative=1, n_clusters_per_class=1, random_state=41)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y,epoch):
  X = np.insert(X,0,1,axis=1)
  weights = np.random.randint(0,1,size=X.shape[1])
  lr = 0.1
  for i in range(epoch):
    j = np.random.randint(0,100)
    z = np.dot(weights,X[j])
    y_hat = step(z)
    weights = weights + lr*(y[j]-y_hat)*X[j]
  return weights[1:], weights[0]

def step(z):
  return 1 if z>0 else 0
```

```python
coef_, intercept_ = perceptron(X,y,2500)
print(coef_)
print(intercept_)
```

```python
# Our eqn is in form w0 + w1*X1 + w2*X2 = 0 --> X2 = -(w1/w2)*X1 -(w0/w2) (y=mx+c form)
# m = -(w1/w2) = -(coeff[0]/coeff[1]) and c = -(intercept/coeff[1])
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
print(m)
print(c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

### Using sigmoid instead of step function:

```python
def perceptron1(X,y,epoch):
  X = np.insert(X,0,1,axis=1)
  weights = np.random.randint(0,1,size=X.shape[1])
  lr = 0.1
  for i in range(epoch):
    j = np.random.randint(0,100)
    z = np.dot(weights,X[j])
    y_hat = sigmoid(z)
    weights = weights + lr*(y[j]-y_hat)*X[j]
  return weights[1:], weights[0]

def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
coef_, intercept_ = perceptron1(X,y,2500)
```

```python
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
```

```python
y_input2 = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input2,color='violet',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.title('Using Sigmoid function',fontsize=20)
plt.ylim(-3,2)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input,y_input1,color='black',linewidth=3)
plt.plot(x_input,y_input2,color='violet',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.title('Comparing 3 different methods',fontsize=20)
plt.ylim(-3,2)
```

We can see that sigmoid function is doing better than the step function as it gives the probability of a point instead of 0 and 1.

### For creating animation of each epoch

```python
def perceptron(X,y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
m,b = perceptron(X,y)
```

```python
fig, ax = plt.subplots(figsize=(9,5))

x_i = np.arange(-3, 3, 0.1)
y_i = x_i*m[0] +b[0]
ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line, = ax.plot(x_i, x_i*m[0] +b[0] , 'r-', linewidth=2)
plt.ylim(-3,3)
def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i*m[i] + b[i])
    ax.set_xlabel(label)
    return line, ax

anim = FuncAnimation(fig, update, repeat=True, frames=200, interval=100)
```
